```python
import wolf
```

```python
import dalmatian
```

```python
pipeline = wolf.ImportTask(
  "git@github.com:getzlab/wolF_characterization_pipeline.git",
  commit = "v1.0.1"
)
```

```python
WORKSPACE = "broad-firecloud-ibmwatson/Brown_Getz_IBM_99-224_Exomes_Liang"
wm = dalmatian.WorkspaceManager(WORKSPACE)
WIC = wolf.fc.WorkspaceInputConnector(WORKSPACE)
P = WIC.get_pairset_as_joint_samples("JB-0372-WGS")
```

```python
with wolf.Workflow(workflow = pipeline.wgs_workflow) as w:
    for pair, p in P.iterrows():
        w.run(
          RUN_NAME = pair,
          t_bam = p["cram_or_bam_path_T"],            # path to tumor BAM (e.g. cloud bucket, or locally stored file)
          t_bai = p["crai_or_bai_path_T"],            # path to tumor BAI
          n_bam = p["cram_or_bam_path_N"],          # path to normal BAM (leave unspecified for tumor-only calling)
          n_bai = p["crai_or_bai_path_N"],         # path to normal BAI (leave unspecified for tumor-only calling)
    
          pairName = pair,          # name of pair
          caseName = pair + "_tumor",        # name of tumor sample (case)
          ctrlName = pair + "_normal",         # name of normal sample (control)
    
          refBuild = 'hg19',        # reference genome build (hg19 or hg38)
          sequencing_type = 'WGS',   # can also be WES
    
          sync = True,              # whether to upload results to Terra workspace
          workspace = WORKSPACE,          # in the format of "<namespace>/<workspace>"
          
          workspace_pairname = pair, # name of Terra entity to attach results to
    
          localization_token = None, # if BAMs are hosted in a location that requires
                                     # an authorization token (e.g. GDC, AWS)
          dry_localization=False,    # run pipeline without localizing BAMs (for
                                     # debugging purposes)

        )
```
